```python
df['Age']= df['Age'].str.replace('_','').str.replace('-500','50')
df['Annual_Income']= df['Annual_Income'].str.replace('_','')
df['Num_of_Loan']= df['Num_of_Loan'].str.replace('_','')
df['Num_of_Delayed_Payment']=df['Num_of_Delayed_Payment'].str.replace('_','')
df['Changed_Credit_Limit']=df['Changed_Credit_Limit'].str.replace('_','')
df['Outstanding_Debt']=df['Outstanding_Debt'].str.replace('_','')
df['Amount_invested_monthly']=df['Amount_invested_monthly'].str.replace('_','')
```

```python
df['Monthly_Balance']=df['Monthly_Balance'].str.replace('_','').str.replace('-','').str.replace('333333333333333333333333333','')
```

```python
df['Age']=pd.to_numeric(df['Age'])
df['Annual_Income']=pd.to_numeric(df['Annual_Income'])
df['Num_of_Loan']=pd.to_numeric(df['Num_of_Loan'])
df['Num_of_Delayed_Payment']=pd.to_numeric(df['Num_of_Delayed_Payment'])
df['Changed_Credit_Limit']=pd.to_numeric(df['Changed_Credit_Limit'])
df['Outstanding_Debt']=pd.to_numeric(df['Outstanding_Debt'])
df['Amount_invested_monthly']=pd.to_numeric(df['Amount_invested_monthly'])
df['Monthly_Balance']=pd.to_numeric(df['Monthly_Balance'])
```

```python
df['Age']= df['Age'].apply(lambda col: np.where(col>100, np.nan, col))
```

```python
df['Before_payment']= df['Delay_from_due_date']
```

```python
df['Num_Bank_Accounts']=df['Num_Bank_Accounts'].apply(lambda col: np.where(col>30, np.nan, col))
df['Num_Credit_Card']=df['Num_Credit_Card'].apply(lambda col: np.where(col>20, np.nan, col))
df['Num_of_Loan']=df['Num_of_Loan'].apply(lambda col: np.where(col<0, np.nan, col))
df['Delay_from_due_date']=df['Delay_from_due_date'].apply(lambda col: np.where(col<0, np.nan, col))
df['Before_payment']=df['Before_payment'].apply(lambda col: np.where(col>0, np.nan, col))
df['Before_payment']=df['Before_payment'].apply(lambda col: np.where(col<0, col*-1, col))
```

```python
df.drop(['ID', 'Customer_ID', 'Name', 'SSN', 'Type_of_Loan', 'Credit_History_Age'], axis=1, inplace=True)
```

```python
if df.Credit_Score.isna().sum()>0:
  df.dropna(subset='Credit_Score')
```

```python
X=df.drop(['Credit_Score'], axis=1)
y=df['Credit_Score'].copy()
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2)
```

```python
num_features =X_train.select_dtypes(include=[np.number]).columns
cat_features =X_train.select_dtypes(exclude=[np.number]).columns
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
num_pipeline= Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

cat_pipeline=Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

transformer= ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
], remainder='passthrough')

estimator=RandomForestClassifier()

full_pipeline=Pipeline([
    ('preprocessing', transformer),
    ('estimator', estimator)
])
```

```python
full_pipeline.fit(X_train, y_train)
```

```python
print(f'Train score:{full_pipeline.score(X_train, y_train)}\nTest score: {full_pipeline.score(X_test, y_test)}')
```
